```python
import pandas as pd
df = pd.read_csv('movie_out.csv', sep='\t', index_col=False)
sdf = df
```

```python
sdf.sort_values('release_year', inplace=True)
sdf
```

```python
sdf = sdf[~sdf['type'].str.contains('Adult')]
sdf
```

```python
sdf = sdf[sdf['runtime'] > 60]
sdf = sdf[sdf['release_year'] <= 2021]
sdf.tail(10000)
```

```python
sdf.to_csv('movie_out.csv', sep='\t', index=False)
#first filter out#
```
